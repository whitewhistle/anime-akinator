```python
import requests
from bs4 import BeautifulSoup

url = "https://www.anime-planet.com/characters/tags"

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

tds = soup.findAll('td', {'class': 'volsChRating'})

for td in tds:
    a_tags = td.findAll('a')
    for tag in a_tags:
        print(tag.text)
```

```python
url2 = "https://www.anime-planet.com/characters/tags?page=2"

response2 = requests.get(url2)
soup2 = BeautifulSoup(response2.text, "html.parser")

tds2 = soup2.find_all("td", class_="volsChRating")

for td2 in tds2:
    a_tags2 = td2.find_all("a")
    for a_tag2 in a_tags2:
        print(a_tag2.text)
```
